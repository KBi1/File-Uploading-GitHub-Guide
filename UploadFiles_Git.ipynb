# **We will used the following commands:**
---
# 1.  `cd desktop`:  By using this command, you will change your current directory to the "desktop" directory.

# 2.  `cd Python_For_Beginners`: This command takes you to the directory "Python_For_Beginners."

# 3.  `git init`: This command creates a new Git repository. It essentially sets up version control for the files in that directory.

# 4. `git status`: Shows you the current status of your repository. Repositories' state is shown by untracked files, changes to be committed, and other details.

image.png

#5. `git add .`: This command saves your changes. You can think of it like staging before saving.

#6. `git status`: This command tells you what's happening in your project right now.




image.png

#7. `git commit -m â€œYour commit message hereâ€`: You can use this command to save your work and add a brief description of what you did. Taking a snapshot of your project and remembering what you changed



image.png

# 8. `git remote add origin <repository URL>`: This command tells Git where to find the remote repository (usually on GitHub). It's like giving Git the address to your project's online storage space.

image.png

# 9. `git push -u origin master`: Replace branch-name with the branch name (usually main or master).
