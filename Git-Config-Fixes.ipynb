# **Set the Configuration**

# In order to set up Git, you will need to use the `git config` command and provide the desired key and value. Please replace "name" (your preferred name) and "email" (your current email address).


## *Set global user.name*
## With git config --global user.name "username," you set the global Git username for your system so that you will use it across all Git repositories. Put your Git username in place of "username".
# `git config --global user.name "username"`



image.png

## *Set global user.email*
##This command sets your global Git email to "emailaddr." You can use it across all your repositories. Replace "emailaddr" with your own address.
# `git config --global user.email "emailaddr"`
image.png

# **Set credential helper to 'store'**
##Git can be configured to store your credentials (like usernames and passwords) to make future authentication easier. This is done by specifying the git config option --global credential.helper 'store'.

#`git config --global credential.helper 'store'`
